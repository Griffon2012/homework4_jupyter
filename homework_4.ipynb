### Задача 1

Исследуйте, какие характеристики недвижимости влияют на стоимость недвижимости, с применением не менее 5 диаграмм из урока.
Анализ сделайте в формате storytelling: дополнить каждый график письменными выводами и наблюдениями.

```python
corr_matrix = df.corr()
corr_matrix = np.round(corr_matrix, 1)
corr_matrix[np.abs(corr_matrix) < 0.3] = 0
corr_matrix

```

```python
plt.figure(figsize=(10, 6))
sns.heatmap(corr_matrix, annot=True, linewidths=.5, cmap='coolwarm');
```

```python
sns.jointplot(x=df['price'], y=df['sqft_living'], kind='reg')
plt.title('Зависимость цены от жилой площади')
plt.ylabel('Жилая площадь (sqft)')
plt.xlabel('Стоимость');
```

```python
sns.jointplot(x=df['price'], y=df['sqft_living'], kind='reg')
plt.title('Зависимость цены от жилой площади')
plt.ylabel('Жилая площадь (sqft)')
plt.xlabel('Стоимость');
```

Прямая зависимость цены жилья от жилой площади

```python
plt.figure(figsize=(16, 8))
sns.boxplot(x=df['price'], y=df['grade'].astype('str'), whis=1.5);
plt.title('Зависимость цены от класса');
```

Прямая зависимость цены от класса жилья.

```python
data = df[(df['sqft_living'] >=1000) & (df['sqft_living'] <=3000)]
corr_matrix = data.corr()
corr_matrix = np.round(corr_matrix, 1)
corr_matrix[np.abs(corr_matrix) < 0.3] = 0
corr_matrix
```

```python
sns.jointplot(x=data['price'], y=data['lat'], kind='kde')
plt.title('Зависимость цены от широты');
```

Если брать основную массу жилья (sqft_living от 1000 до 3000) то прослеживается прямая зависимость цены от широты (чем севернее, тем дороже)

```python
data = df[(df['sqft_living'] > 3000)]
corr_matrix = data.corr()
corr_matrix = np.round(corr_matrix, 1)
corr_matrix[np.abs(corr_matrix) < 0.3] = 0
corr_matrix
```

```python
plt.figure(figsize=(16, 8))
sns.boxplot(x=data['price'], y=data['view'].astype('str'), whis=1.5);
plt.title('Зависимость цены вида');
```

На жилье квадратурой жилой площади более 3000 начинается сказывается вид неа набережную и общий вид на жилье. Так же прямая зависимость от количества ванных комнат.
